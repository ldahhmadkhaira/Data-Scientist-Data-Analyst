# Performance Monitoring System Analysis
Analisis mendalam terhadap kinerja perusahaan berdasarkan dataset Superstore (2012-2015).

```
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
from datetime import datetime
import numpy as np

# Load processed data
df = pd.read_csv('cleaned_superstore.csv')
monthly = pd.read_csv('monthly_kpi.csv')
df['Order Date'] = pd.to_datetime(df['Order Date'])
```

## 1. KPI Achievement Analysis
Menganalisis pencapaian target penjualan bulanan. Target ditetapkan sebesar 115% dari penjualan bulan yang sama pada tahun sebelumnya.

```
fig = go.Figure()
fig.add_trace(go.Scatter(x=monthly['Date'], y=monthly['Sales'], name='Actual Sales', line=dict(color='cyan', width=3)))
fig.add_trace(go.Scatter(x=monthly['Date'], y=monthly['Target_Sales'], name='Target Sales', line=dict(color='orange', dash='dash')))
fig.update_layout(title='Target vs Actual Sales Revenue', template='plotly_dark')
fig.show()
```

## 2. Analisis Performa per Unit (Category & Market)
Mengidentifikasi divisi mana yang paling bermasalah.

```
# Hitung profit margin per category
cat_perf = df.groupby('Category').agg({'Sales':'sum', 'Profit':'sum'}).reset_index()
cat_perf['Margin'] = (cat_perf['Profit'] / cat_perf['Sales']) * 100
fig_cat = px.bar(cat_perf, x='Category', y='Sales', color='Margin', 
                 title='Sales by Category with Profit Margin', template='plotly_dark')
fig_cat.show()
```

## 3. Analisis Geografis (Market)
Pasar mana yang memberikan kontribusi pertumbuhan terbesar vs yang tertinggal.

```
market_perf = df.groupby('Market').agg({'Sales':'sum', 'Profit':'sum'}).reset_index()
fig_market = px.pie(market_perf, values='Sales', names='Market', 
                     title='Market Share Distribution', hole=.3, template='plotly_dark')
fig_market.show()
```

## 4. Insight & Rekomendasi Bisnis
### Temuan Utama:
1. **Trend Penjualan**: Terdapat fluktuasi musiman yang kuat, terutama lonjakan pada akhir tahun (Q4).
2. **Divisi Bermasalah**: Furniture memiliki profit margin yang jauh lebih rendah dibandingkan Technology & Office Supplies, meskipun sales-nya cukup tinggi.
3. **KPI Achievement**: Teridentifikasi beberapa bulan di mana target 115% gagal tercapai (Achievement < 100%).

### Rekomendasi:
- **Optimalisasi Divisi Furniture**: Tinjau ulang biaya operasional dan diskon, karena meskipun sales besar, kontribusi profitnya rendah.
- **Inventory Planning**: Siapkan stok lebih awal untuk Q4 guna menangkap lonjakan permintaan.
